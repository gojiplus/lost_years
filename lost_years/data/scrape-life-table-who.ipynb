```python
import pandas as pd
import requests
import pycountry
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(requests.get('https://apps.who.int/gho/data/node.main.LIFECOUNTRY?lang=en').text, 'lxml')
```

```python
countries = {'Afghanistan': 'AFG'}
```

```python
base_url = 'https://apps.who.int/gho/data/'
for a in soup.select('ul.list_dash li a'):
    print(a.text)
    url = base_url + a['href']
    soup2 = BeautifulSoup(requests.get(url).text, 'lxml')
    for iframe in soup2.find_all('iframe'):
        src = iframe['src']
        break
    countries[a.text] = src.split(':')[-1]
countries
```

```python
from io import StringIO

adf = pd.DataFrame()
```

```python
skip = True
for c in countries:
    print(c)
    if c == 'Romania':
        skip = False
        print(countries[c])
    if skip:
        continue
    url = 'https://apps.who.int/gho/athena/data/data-verbose.csv?target=GHO/LIFE_0000000029,LIFE_0000000030,LIFE_0000000031,LIFE_0000000032,LIFE_0000000033,LIFE_0000000034,LIFE_0000000035&profile=verbose&filter=COUNTRY:%s' % countries[c]
    r = requests.get(url)
    if r.status_code == 200:
        try:
            df = pd.read_csv(StringIO(r.text))
            adf = adf.append(df)
        except Exception as e:
            print(e)
    else:
        print('Error %s' % c)
```
